```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
names = ['hid_16_3layers', 'hid_16_bid', 'hid_16', 'hid_32', 'hid_16_1layer', 'hid_8']
```

```python
batch_nums = np.arange(134) * 20
```

```python
colors = ['blue', 'green', 'red', 'orange', 'cyan', 'magenta']
```

```python
descs = ['16 neurons, 3 layers', '16 neurons, 2 layers, biderectional', '16 neurons, 2 layers', 
         '32 neurons, 2 layers', '16 neurons, 1 layer', '8 neurons, 2 layers']
```

```python
plt.figure(figsize=(12, 9))
for name, color, desc in zip(names, colors, descs):
    train_history = np.loadtxt('data/' + name + '_train.txt')
    val_history = np.loadtxt('data/' + name + '_val.txt')
    
    # label=f'{name}
    batch_nums = np.arange(len(train_history)) * 20
    plt.plot(batch_nums, train_history, c=color, alpha=0.5, lw=2, linestyle='-.', label=desc)
    plt.plot(batch_nums, val_history, c=color, alpha=1, lw=5)

plt.ylim(2, 5)
plt.title('Learning curves for LSTMs', fontsize=16)
plt.ylabel('loss', fontsize=16)
plt.xlabel('n_batch', fontsize=16)
plt.legend(fontsize=14)
plt.tight_layout()
plt.savefig('LSTM_new.png')
```
